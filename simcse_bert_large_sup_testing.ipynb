```python
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score, f1_score
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import multilabel_confusion_matrix

# Set the threshold for classification
thr = 0.5

# Define the dataset class
class CustomDataset(Dataset):
    def __init__(self, data, tokenizer, max_length):
        self.data = data.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.loc[idx, 'TEXT']
        labels = self.data.loc[idx, ['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN']].values
        labels = [1 if label == 'y' else 0 for label in labels]

        input_ids, attention_mask = preprocess_text(text, self.tokenizer, self.max_length)

        return {
            'input_ids': torch.tensor(input_ids, dtype=torch.long),
            'attention_mask': torch.tensor(attention_mask, dtype=torch.long),
            'labels': torch.tensor(labels, dtype=torch.float)
        }

def preprocess_text(text, tokenizer, max_length):
    cleaned_text = text.lower().strip()
    tokens = tokenizer.tokenize(cleaned_text)
    tokens = ['[CLS]'] + tokens + ['[SEP]']

    # Account for the special tokens added to the tokens list
    special_tokens_count = 2  # Number of special tokens: [CLS] and [SEP]

    # Truncate tokens if needed
    if len(tokens) > max_length - special_tokens_count:
        tokens = tokens[: max_length - special_tokens_count]

    input_ids = tokenizer.convert_tokens_to_ids(tokens)

    # Pad sequence if needed
    padding_length = max_length - len(input_ids) - special_tokens_count
    input_ids = [tokenizer.pad_token_id] * padding_length + input_ids + [tokenizer.pad_token_id]

    attention_mask = [1] * len(input_ids)

    return input_ids, attention_mask

model_name = 'princeton-nlp/sup-simcse-bert-large-uncased'
# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained(model_name)

# Define the maximum sequence length
max_length = 256

# Create the dataset
dataset = CustomDataset(df, tokenizer, max_length)

# Split the dataset into train and test sets
label_data = df[['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN']].apply(lambda x: ' '.join(x), axis=1)
stratified_split = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
train_indices, test_indices = next(stratified_split.split(df, label_data))

train_data = torch.utils.data.Subset(dataset, train_indices)
test_data = torch.utils.data.Subset(dataset, test_indices)

print(len(train_data), len(test_data))

# Define the batch size for training and evaluation
batch_size = 24

# Create the data loaders
train_dataloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=batch_size)

# Get the number of unique labels
num_labels = dataset.data[['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN']].values.shape[1]
print(f"Number of unique labels: {num_labels}")

# Load the BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

# Define the optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)

# Define the learning rate scheduler
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.1, patience=3)

# Define the early stopping criteria
class EarlyStopping:
    def __init__(self, patience=5):
        self.patience = patience
        self.counter = 0
        self.best_score = None
        self.early_stop = False

    def __call__(self, score):
        if self.best_score is None:
            self.best_score = score
        elif score < self.best_score:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.counter = 0
        return self.early_stop

early_stopping = EarlyStopping(patience=3)

# Train the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

num_epochs = 30
total_steps = len(train_dataloader) * num_epochs

# Set the model to training mode
model.train()

scaler = torch.cuda.amp.GradScaler()

best_f1 = 0.0
best_model_path = 'best_model.pt'

train_losses = []
val_losses = []
val_accuracies = []
val_aucs = []
val_f1s = []
val_cm = []

for epoch in range(num_epochs):
    total_loss = 0
    total_train_correct = 0
    total_train_samples = 0
    total_val_loss = 0

    for batch in train_dataloader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)

        loss = outputs.loss
        logits = outputs.logits

        scaler.scale(loss).backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # Optional gradient clipping
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

        predicted_labels = (logits.sigmoid() > thr).float()
        total_train_correct += (predicted_labels == labels).sum().item()
        total_train_samples += labels.size(0)

    average_train_loss = total_loss / len(train_dataloader)
    train_accuracy = total_train_correct / (total_train_samples * num_labels)

    train_losses.append(average_train_loss)

    print(f"Epoch {epoch + 1}/{num_epochs}")
    print(f"Train Loss: {average_train_loss:.4f} | Train Accuracy: {train_accuracy:.4f}")

    model.eval()

    with torch.no_grad():
        total_val_correct = 0
        total_val_samples = 0
        all_predicted_labels = []
        all_true_labels = []

        for batch in test_dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)

            val_loss = outputs.loss
            logits = outputs.logits
            total_val_loss += val_loss.item()
            predicted_labels = (logits.sigmoid() > thr).float()
            total_val_correct += (predicted_labels == labels).sum().item()
            total_val_samples += labels.size(0)

            all_predicted_labels.extend(predicted_labels.cpu().tolist())
            all_true_labels.extend(labels.cpu().tolist())

        average_val_loss = total_val_loss / len(test_dataloader)
        val_accuracy = total_val_correct / (total_val_samples * num_labels)

        val_auc = roc_auc_score(all_true_labels, all_predicted_labels, average='macro', multi_class='ovr')
        val_f1 = f1_score(all_true_labels, all_predicted_labels, average='macro')
        val_confusion_matrix = multilabel_confusion_matrix(all_true_labels, all_predicted_labels)

        val_losses.append(average_val_loss)
        val_accuracies.append(val_accuracy)
        val_aucs.append(val_auc)
        val_f1s.append(val_f1)
        val_cm.append(val_confusion_matrix)

        print(f"Val Loss: {average_val_loss:.4f} | Val Accuracy: {val_accuracy:.4f}")
        print(f"Val AUC: {val_auc:.4f} | Val F1 Score: {val_f1:.4f}")

        if val_f1 > best_f1:
            best_f1 = val_f1
            torch.save(model.state_dict(), best_model_path)
            print("..............Saved best model!................")

        scheduler.step(val_f1)

        if early_stopping(val_f1):
            print("Early stopping triggered!")
            break

    model.train()


```

```python
# Generate and save graphs
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

def generate_and_save_graphs(train_losses, val_losses, val_accuracies, val_aucs, val_f1s, val_cm, save_dir):
    # Create save directory if it doesn't exist
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    epochs = range(1, len(train_losses) + 1)

    # Training and Validation Loss Curve
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, train_losses, label='Training Loss')
    plt.plot(epochs, val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.savefig(f'{save_dir}/loss_curve.png')
    plt.close()

    # Training and Validation Accuracy Curve
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, val_accuracies)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Validation Accuracy')
    plt.savefig(f'{save_dir}/accuracy_curve.png')
    plt.close()

    # Plot and save train losses
    plt.figure()
    plt.plot(train_losses)
    plt.xlabel('Epoch')
    plt.ylabel('Train Loss')
    plt.title('Train Loss over Epochs')
    plt.savefig(os.path.join(save_dir, 'train_losses.png'))
    plt.close()

    # Plot and save validation losses
    plt.figure()
    plt.plot(val_losses)
    plt.xlabel('Epoch')
    plt.ylabel('Validation Loss')
    plt.title('Validation Loss over Epochs')
    plt.savefig(os.path.join(save_dir, 'val_losses.png'))
    plt.close()

    # Plot and save validation accuracies
    plt.figure()
    plt.plot(val_accuracies)
    plt.xlabel('Epoch')
    plt.ylabel('Validation Accuracy')
    plt.title('Validation Accuracy over Epochs')
    plt.savefig(os.path.join(save_dir, 'val_accuracies.png'))
    plt.close()

    # Plot and save validation AUCs
    plt.figure()
    plt.plot(val_aucs)
    plt.xlabel('Epoch')
    plt.ylabel('Validation AUC')
    plt.title('Validation AUC over Epochs')
    plt.savefig(os.path.join(save_dir, 'val_aucs.png'))
    plt.close()

    # Plot and save validation F1 scores
    plt.figure()
    plt.plot(val_f1s)
    plt.xlabel('Epoch')
    plt.ylabel('Validation F1 Score')
    plt.title('Validation F1 Score over Epochs')
    plt.savefig(os.path.join(save_dir, 'val_f1s.png'))
    plt.close()

    # Flatten and concatenate confusion matrices
    flat_cm = np.concatenate([cm.reshape(1, -1) for cm in val_cm], axis=0)

    # Plot confusion matrix
    plt.figure(figsize=(10, 5))
    sns.heatmap(flat_cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.savefig(os.path.join(save_dir, 'confusion_matrix.png'))
    plt.close()


save_dir = './'
generate_and_save_graphs(train_losses, val_losses, val_accuracies, val_aucs, val_f1s, val_cm, save_dir)
```

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score, f1_score, multilabel_confusion_matrix
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import DataLoader, Dataset

# Set the threshold for classification
thr = 0.5

# Define the dataset class
class CustomDataset(Dataset):
    def __init__(self, data, tokenizer, max_length):
        self.data = data.reset_index(drop=True)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.loc[idx, 'TEXT']
        labels = self.data.loc[idx, ['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN']].values
        labels = [1 if label == 'y' else 0 for label in labels]

        input_ids, attention_mask = preprocess_text(text, self.tokenizer, self.max_length)

        return {
            'input_ids': torch.tensor(input_ids, dtype=torch.long),
            'attention_mask': torch.tensor(attention_mask, dtype=torch.long),
            'labels': torch.tensor(labels, dtype=torch.float)
        }

def preprocess_text(text, tokenizer, max_length):
    cleaned_text = text.lower().strip()
    tokens = tokenizer.tokenize(cleaned_text)
    tokens = ['[CLS]'] + tokens + ['[SEP]']

    # Account for the special tokens added to the tokens list
    special_tokens_count = 2  # Number of special tokens: [CLS] and [SEP]

    # Truncate tokens if needed
    if len(tokens) > max_length - special_tokens_count:
        tokens = tokens[: max_length - special_tokens_count]

    input_ids = tokenizer.convert_tokens_to_ids(tokens)

    # Pad sequence if needed
    padding_length = max_length - len(input_ids) - special_tokens_count
    input_ids = [tokenizer.pad_token_id] * padding_length + input_ids + [tokenizer.pad_token_id]

    attention_mask = [1] * len(input_ids)

    return input_ids, attention_mask

# Load the pre-trained BERT model
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=5)

# Load the saved model weights
model.load_state_dict(torch.load("best_model.pt"))

# Set the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Load the test data
testing_path = '/content/SOPs.xlsx - Sheet1.csv'
test_df = pd.read_csv(testing_path)

# Create the test dataset
test_dataset = CustomDataset(test_df, tokenizer, max_length=256)

# Create the test data loader
batch_size = 64
test_dataloader = DataLoader(test_dataset, batch_size=batch_size)

# Set the model to evaluation mode
model.eval()

# Make predictions on the test data
all_predicted_labels = []
with torch.no_grad():
    for batch in test_dataloader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        predicted_labels = (logits.sigmoid() > thr).float()
        all_predicted_labels.extend(predicted_labels.cpu().tolist())

# Convert the predicted labels to a DataFrame
predicted_labels_df = pd.DataFrame(all_predicted_labels, columns=['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN'])
output_df = pd.concat([test_df['TEXT'], predicted_labels_df], axis=1)

# Save the output DataFrame as a CSV file
output_df.to_csv('predictions.csv', index=False)
```
